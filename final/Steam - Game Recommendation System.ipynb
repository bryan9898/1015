## Content-Based Recommendation
We will build a system that recommends games with features that are similar to a particular game, based on the contents of that particular game itself (item metadata). 
- We will use text data containing `short_description`, `genres`, `additional_tags`, `developer`, `publisher`, `platforms` of the game.
- We will then use `CountVectorizer`, which will vectorize the text data, while not down-weighting developer/publishers. We will then find the pairwise **cosine similarity scores** for that particular game with all games, then output the top most similar games as recommendations.

```python
data = pd.read_csv("steamdata_clean.csv")
```

```python
import re

def clean_text(sample):
    if sample is np.nan:
        return ""
    
    sample = sample.lower()
    
    # remove formatting, punctuation
    sample = sample.replace("<br /><br />", "")
    sample = re.sub("[^a-zA-Z]+", " ", sample)
    
    sample = sample.split()
    
    sample = " ".join(sample)
    
    return sample

def join_tags(sample):
    if sample is np.nan:
        return ""
    return sample.replace(";", " ")

def join_tags_removespacing(sample):
    if sample is np.nan:
        return ""
    tags_list = sample.split(";")
    for tag in range(len(tags_list)):
        tags_list[tag] = tags_list[tag].replace(" ", "")
    return " ".join(tags_list)

def join_text_variables(df):
    short_description = df["short_description"].apply(clean_text)
    additional_tags = df["additional_tags"].apply(join_tags)
    genres = df["genres"].apply(join_tags)
    developer = df["developer"].apply(join_tags_removespacing)
    publisher = df["publisher"].apply(join_tags_removespacing)
    platforms = df["platforms"].apply(join_tags)

    return short_description + " " + additional_tags + " " + genres + " " + developer + " " + publisher + " " + platforms
```

```python
# clean text data, then join text variables of description + additional tags + genres + developers + publishers + platforms
data["joined"] = join_text_variables(data)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count = CountVectorizer(stop_words = "english")

# creates matrix where each column represents a word in the overview vocabulary
count_matrix = count.fit_transform(data["joined"])
count_matrix
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# calculate pairwise cosine similarities
cosine_sim = cosine_similarity(count_matrix, count_matrix)
```

```python
# output top N similar games using name of game and cosine similarities
def get_recommendations(df, cosine_sim, name, N):
    if name in df["name"].values:
        # get the dataset index of the game that exactly matches the input name
        game_index = df.loc[df["name"] == name].index[0]

        # get the pairwise similarity scores of all games with that game
        sim_scores = pd.DataFrame(list(enumerate(cosine_sim[game_index]))).sort_values(1, ascending = False)
        # select top N similar games
        top_N = sim_scores.head(N+1).reset_index(drop = True)

        # replacing dataset index with game names for top N similar games
        for game in range(1, N+1):
            i = top_N[0][game]
            top_N.loc[game, 0] = df.iloc[i]["name"]
        top_N.columns = ["name", "similarity"]

        return top_N[1:]
    else:
        return "Game does not exist in our database!"
```

```python
game_name = input("Enter name of your favourite game: ")
print(f"Since you liked {game_name}, you should also try:")
print(get_recommendations(data, cosine_sim, game_name, 10))
```

```python
userData = pd.read_csv("playergames.csv")

rating_crosstab = userData.pivot_table(values='rating', index='steamid', columns='gamesid', fill_value=0)
rating_crosstab.columns = rating_crosstab.columns.astype(int)
rating_crosstab
```

```python
# transpose the matrix
X = rating_crosstab.T
```

```python
from sklearn.decomposition import TruncatedSVD

SVD = TruncatedSVD(n_components=400)
result_matrix = SVD.fit_transform(X)    
print(f"Explained Variance of Truncated SVD: {sum(SVD.explained_variance_ratio_):3f}")
```

```python
from sklearn.manifold import TSNE

tsne = TSNE(perplexity=5, n_components=2, init="pca", n_iter=5000, learning_rate = "auto")
plot_only = 50
coords = tsne.fit_transform(result_matrix[:plot_only, :])

plt.figure(figsize=(18, 18))
labels = [i for i in range(1, 50)]
for i, label in enumerate(labels):
    x, y = coords[i, :]
    plt.scatter(x, y)
    plt.annotate(label,
                 xy=(x, y),
                 xytext=(10, 4),
                 textcoords="offset points",
                 ha="right",
                 va="bottom")

plt.show()
```

With t-SNE, We can see that SVD is finding points close to each other within different dimensions and grouping them up.

Next, we want to find out how similar each game is to other games based off a user's ratings. For each game, we will calculate their similarities (using Pearson's correlation coefficient), based on the user's perspective.

```python
corr_matrix = np.corrcoef(result_matrix)
```

```python
def get_fav_games(df):
    fav_games = dict()
    while True:
        name = input("Enter name of your favourite game (eg. Counter-Strike) (input -1 to stop): ")
        
        if name == '-1':
            return fav_games
        
        elif name in df["name"].values:
            rating = input(f"Enter your rating of {name} from 0 to 5 (eg. 5): ")
            if not rating.isdigit() or int(rating) < 0 or int(rating) > 5:
                return "Invalid rating!"
            fav_games[name] = rating
        else:
            return "Game does not exist in our database!"
```

```python
fav_games = get_fav_games(data)
```

```python
fav_games
```

```python
# output top N similar games using name of game and correlation matrix
def get_recommendations_corr(df, corr_matrix, rating_crosstab, name, N):
    if name in df["name"].values:
        # get the dataset index of the game that exactly matches the input name
        game_index = df["steam_appid"].loc[data["name"] == name]
        
        # list of all game IDs from userData
        gameid_list = list(rating_crosstab.columns)

        # position of game in UserData
        game_position = gameid_list.index(game_index.iloc[0])
        
        # isolating the array that represents games played
        corr_game = corr_matrix[game_position]
        
        # get correlations of all games with that game
        all_corr = pd.DataFrame(corr_game).reset_index().sort_values(0, ascending = False)

        # select top N similar games 
        top_N = all_corr.head(N+1).reset_index(drop = True)

        # replacing dataset index with game names for top N similar games
        for game in range(1, N+1):
            i = top_N["index"][game]
            top_N.loc[game, "index"] = data[data["steam_appid"] == gameid_list[i]]["name"].iloc[0]
        top_N.columns = ["name", "correlation"]

        return(top_N[1:])

    else:
        return "Game does not exist in our database!"
```

```python
rec = pd.DataFrame()
if len(fav_games) < 3:
    for game_name in fav_games:
        rec = pd.concat([rec, get_recommendations_corr(data, corr_matrix, rating_crosstab, game_name, 10//len(fav_games))], axis = 0)
else:
    fav_games = sorted(fav_games, key=fav_games.get, reverse=True)[:3]
    for game_name in fav_games:
        rec = pd.concat([rec, get_recommendations_corr(data, corr_matrix, rating_crosstab, game_name, 3)], axis = 0)

# remove repeated games and print recommendations
print(f"Since you liked {fav_games}, you should also try:")        
print(rec.drop_duplicates("name").reset_index(drop = True))
```
